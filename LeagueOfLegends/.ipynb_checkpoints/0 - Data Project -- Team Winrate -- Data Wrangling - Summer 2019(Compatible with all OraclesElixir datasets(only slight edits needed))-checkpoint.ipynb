### Import Data

```python
infile = r"C:\Users\Triplea657\000 MSCS-335 2020\Datasets\League_2019\2019-worlds-"
inf = "OraclesElixir.csv"
outf = "Wrangled.csv"
alldata = pd.read_csv(infile+inf,low_memory=False)
```

```python
#find data length for operations
alldata.count(axis=0)[0]
```

### Grab data that we want (Team statistics)

```python
inddata = alldata
for i in range(int((inddata.count(axis=0)[0])/12)):#drop player rows, leaving only team stats
    for j in range(10):
        if ((j != 10) and (j != 11)):
            inddata = inddata.drop([12*(i)+j])

# make sure the operations worked properly
print(inddata.head(15))
inddata.tail(21)
```

```python
var = []
for i in inddata:
    var.append(i)
print(var)
for i, v in enumerate(var):
    print(i, v)
```

```python
data = inddata.copy()
#del inddata

#add desired column numbers to list
tokeep = [2]
for i in range(5):
    tokeep.append(19+i)
tokeep.append(30)
for i in range(20):
    tokeep.append(34+i)
for i in range(6):
    tokeep.append(57+i)
tokeep.append(65)
tokeep.append(66)
tokeep.append(68)
tokeep.append(69)
for i in range(4):
    tokeep.append(74+i)
for i in range(17):
    tokeep.append(80+i)


print("keeping column IDs:{}".format(tokeep))
tokeepstr = ""
for i in tokeep:
    if i != tokeep[0]:
        tokeepstr += ", " 
    tokeepstr += var[i]
print("keeping columns: {}".format(tokeepstr))

datakp = data.iloc[:,tokeep]
datakp = datakp[datakp['league']!='LPL']#had to drop chinese games since they keep so little data
print("Kept data:")
datakp.head()
```

```python
#datakp = datakp.reset_index(drop=True,inplace=True) #reindex data with new indices
#print(datakp.head())
#datakp = datakp.drop(['index'])

hotleague = pd.get_dummies(datakp["league"],prefix='league')
dataleague = datakp.drop(['league'],axis=1)
hotdata = pd.concat([hotleague,dataleague],axis=1)
hotdata = hotdata.astype(float)
hotdata.head()
```

```python
#write 
hotdata.to_csv(infile+outf)
```
