## 1 Wrangling

### 1.1 import data

```python
alldata = pd.read_csv(r"StarCraftData.csv",low_memory=False)
```

```python
datalength = alldata.count(axis=0)[0]
print(datalength)
alldata.head()
```

### 1.2 Choose desired data and add onehot encoding for league

```python
alldata = alldata.drop('GameID',1) #remove GameID
alldata.head()
```

```python
hotleague = pd.get_dummies(alldata["LeagueIndex"])
leagues = ["Bronze","Silver","Gold","Platinum","Diamond","Master","GrandMaster"]
for i in hotleague.columns:
    hotleague = hotleague.rename(columns={i:leagues[i-1]})
hotdata = pd.concat([hotleague,alldata],axis=1)
hotdata = hotdata.astype(float)
hotdata.head(10)
```

### 1.3 examine data

```python
hotdata = hotdata.astype(np.float32, copy=True, errors='raise')
hotdata.info()
```

```python
#check for empty
print(np.where(pd.isnull(hotdata)))
for i,j in pd.isna(hotdata).iteritems():
    if j.any() == True:
        print(j)
```

```python
cols = []
for i in hotdata:
    cols.append(i)
cols
```

```python
n=len(hotdata[cols[0]])
print(f"{n}")
print(hotdata[cols[0]][n-1])
print()
```

## Make datasets in which Leagues are grouped for easier differentiation

#### groupings chosen based on data analysis in part 1 - Data Comprehension

```python
Elos = pd.DataFrame(np.zeros((len(alldata),4)))
for i, val in enumerate(Elos.values):
    if (alldata["LeagueIndex"][i] < 4):
        Elos[0][i] = 1.0
        Elos[3][i] = 0.0
    elif(3 < alldata["LeagueIndex"][i] and alldata["LeagueIndex"][i] < 6):
        Elos[1][i] = 1.0
        Elos[3][i] = 1.0
    elif(5 < alldata["LeagueIndex"][i] and alldata["LeagueIndex"][i] < 8):
        Elos[2][i] = 1.0
        Elos[3][i] = 2.0
Elos = Elos.rename(columns={0:"Low_ELO",1:"Med_ELO",2:"High_ELO",3:"GroupedLeagueIndex"})
Elos.head()
hotdata2 = pd.concat([Elos,alldata],axis=1)
hotdata2 = hotdata2.astype(float)
hotdata2.head(10)
```

```python
hotdata2.to_csv("Data-Prepped-ELO-3-Groups.csv")
```

```python
Elos = pd.DataFrame(np.zeros((len(alldata),3)))
for i, val in enumerate(Elos.values):
    if (alldata["LeagueIndex"][i] < 4):
        Elos[0][i] = 1.0
        Elos[2][i] = 0.0
    elif(4 < alldata["LeagueIndex"][i]):
        Elos[1][i] = 1.0
        Elos[2][i] = 1.0
Elos = Elos.rename(columns={0:"Low_ELO",1:"High_ELO",2:"GroupedLeagueIndex"})
Elos.head()
hotdata3 = pd.concat([Elos,alldata],axis=1)
hotdata3 = hotdata3.astype(float)
hotdata3.head(10)
```

```python
hotdata3.to_csv("Data-Prepped-ELO-2-Groups.csv")
```
